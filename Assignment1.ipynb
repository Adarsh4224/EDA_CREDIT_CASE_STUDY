```python
#imbalance percentage
print(100*d1.TARGET.value_counts()/ len(d1))
plt.show((d1.TARGET.value_counts()/ len(d1)).plot.bar())
# clearly imbalance of 91.9% to 8.o7% in Target variable i.e.payment difficulties as expected imbalance.
Target0 = d1.loc[d1["TARGET"]==0] 
Target1 = d1.loc[d1["TARGET"]==1]
print('Imbalance percentage')
print(round(len(Target0)/len(Target1),2))
```

```python
#now performing univariant analysis
Target0.WEEKDAY_APPR_PROCESS_START.value_counts(normalize=True).plot.bar()
plt.title('non-default')
plt.show()
Target1.WEEKDAY_APPR_PROCESS_START.value_counts(normalize=True).plot.bar()
plt.title('default')
plt.show()
# conclusiion: 1. less default and non default on weekends means actually  there are less applications on weekends.
```

```python
# for categorical columns
list_categorical_columns=['NAME_CONTRACT_TYPE','CODE_GENDER','FLAG_OWN_CAR','FLAG_OWN_REALTY',
                     'NAME_EDUCATION_TYPE','NAME_FAMILY_STATUS','NAME_HOUSING_TYPE',
                    'WEEKDAY_APPR_PROCESS_START','AGE_BRACKET','AMT_INCOME_TOTAL_bin']
plt.figure(figsize=(22,25))
for i in list_categorical_columns:
    print(i)
    Target0[i].value_counts(normalize=True).plot.bar()
    plt.title('non-default')
    plt.show()
    Target1[i].value_counts(normalize=True).plot.bar()
    plt.title('default')
    plt.show()
    
#Looking at graphs following insights are apparent.
#people who have high credit amount,applied on weekends and higher education are less likely to default.
#  people with car loan, appartment  and secondary/special educated people are applying loans in high in number
```

```python
# for continous numerical columns.
list_continous_numerical_columns=['AMT_ANNUITY','AMT_GOODS_PRICE','CNT_FAM_MEMBERS','DAYS_LAST_PHONE_CHANGE','DAYS_ID_PUBLISH','DAYS_BIRTH','HOUR_APPR_PROCESS_START','DAYS_EMPLOYED','AMT_CREDIT','AMT_INCOME_TOTAL']
plt.figure(figsize=(22,25))
for i in (enumerate(list_continous_numerical_columns)):
    plt.subplot(len(list_continous_numerical_columns)//2,2,i[0]+1)
    sns.distplot(Target0[i[1]].dropna(),hist=False,label='Target : default')
    sns.distplot(Target1[i[1]].dropna(),hist=False,label='Target : no default')
plt.show()    
#Looking at graphs following insights are apparent.
#people who take more loans are  just got employed, retired,age between 27-41yrs,nuclear family and low amount annuity
    
```

```python
#correcting the negatives problem like the above column.
d2['DAYS_DECISION']=d2['DAYS_DECISION'].abs()
d2['SELLERPLACE_AREA']=d2['SELLERPLACE_AREA'].abs()
d2['DAYS_FIRST_DUE']=d2['DAYS_FIRST_DUE'].abs()
d2['DAYS_LAST_DUE_1ST_VERSION']=d2['DAYS_LAST_DUE_1ST_VERSION'].abs()
d2['DAYS_LAST_DUE']=d2['DAYS_LAST_DUE'].abs()
d2['DAYS_TERMINATION']=d2['DAYS_TERMINATION'].abs()
d2['DAYS_FIRST_DRAWING']=d2['DAYS_FIRST_DRAWING'].abs()
```

```python
#mearging d1 and d2
merged=pd.merge(d1,d2,how='left',on='SK_ID_CURR',suffixes=('_Current', '_Previous'))
merged.head()
```

```python
# continuing univarent analysis(catagorical)
categorical=['NAME_CONTRACT_TYPE_Current','NAME_CONTRACT_TYPE_Previous','NAME_TYPE_SUITE_Current','NAME_TYPE_SUITE_Previous','WEEKDAY_APPR_PROCESS_START_Current','WEEKDAY_APPR_PROCESS_START_Previous','NAME_YIELD_GROUP','NAME_CLIENT_TYPE']
plt.figure(figsize=(22,25))
for i in (enumerate(categorical)):
    plt.subplot(len(categorical)//2,2,i[0]+1)
    sns.countplot(x=i[1],hue='NAME_CONTRACT_STATUS',data=merged)
plt.show()
```

```python
# continious numerical analysis
continous=['AMT_CREDIT_Previous','AMT_CREDIT_Current','AMT_ANNUITY_Current','AMT_ANNUITY_Previous','AMT_GOODS_PRICE_Current','AMT_GOODS_PRICE_Previous','CNT_FAM_MEMBERS','CNT_CHILDREN','HOUR_APPR_PROCESS_START_Previous','HOUR_APPR_PROCESS_START_Current']
plt.figure(figsize=(22,25))
for i in (enumerate(continous)):
    plt.subplot(len(continous)//2,2,i[0]+1)
    sns.distplot(merged.loc[merged.NAME_CONTRACT_STATUS=='Approved',:][i[1]].dropna(),hist=False,label='Approved')
    sns.distplot(merged.loc[merged.NAME_CONTRACT_STATUS=='Canceled',:][i[1]].dropna(),hist=False,label='Canceled',kde_kws={'bw':0.1})
    sns.distplot(merged.loc[merged.NAME_CONTRACT_STATUS=='Refused',:][i[1]].dropna(),hist=False,label='Refused',kde_kws={'bw':0.1})
    sns.distplot(merged.loc[merged.NAME_CONTRACT_STATUS=='Unused offer',:][i[1]].dropna(),hist=False,label='Unused offer')

plt.show()
```

```python
#bivarient Analysis
#categorical
for i in ['NAME_CONTRACT_TYPE_Current','NAME_CONTRACT_TYPE_Previous','NAME_TYPE_SUITE_Current','NAME_TYPE_SUITE_Previous','WEEKDAY_APPR_PROCESS_START_Current','WEEKDAY_APPR_PROCESS_START_Previous','NAME_YIELD_GROUP','NAME_CLIENT_TYPE']:
    print(pd.crosstab(index=merged['NAME_CONTRACT_STATUS'],columns=merged[i]))
    (pd.crosstab(index=merged['NAME_CONTRACT_STATUS'],columns=merged[i])).plot(kind="bar", figsize=(5,5),stacked=False)
    plt.show()
```
